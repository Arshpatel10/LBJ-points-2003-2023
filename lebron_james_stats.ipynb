```python
lebron = pd.read_csv("Lebron_James_All_Regular_Season_Stats.csv")
```

```python
lebron.info()
```

```python
lebron.shape
#1421 rows, 29 columns
```

```python
lebron['Location'] = lebron['Location'].replace({'@':'AWAY'}).fillna('HOME')
lebron
```

```python
lebron['Margin'] = lebron['Result'].str.extract(r'\(([-+]?\d+)\)')
lebron['Result'] = lebron['Result'].str.extract(r'([LW])')
lebron.rename(columns = {'Tm': 'Team'}, inplace=True)
lebron
```

```python
lebron.to_excel('lebron_james_regular_season_games_final.xlsx', sheet_name = 'Data')
```
